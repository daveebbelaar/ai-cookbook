```python
import asyncio
from mcp.client.sse import sse_client
from mcp.client.session import ClientSession

# Ensure the server is running in a separate terminal:
# python 4.5-langchain-interation/server.py

async def run_client():
    # Connect to the SSE server
    async with sse_client("http://localhost:4567/sse") as streams:
        async with ClientSession(streams[0], streams[1]) as session:
            # Initialize connection
            await session.initialize()
            
            # List available tools
            tools = await session.list_tools()
            print(f"Connected to server. Available tools: {[tool.name for tool in tools.tools]}")
            
            # Call the 'get_state_and_qrid' tool
            # Example coordinates for Seattle, WA
            result = await session.call_tool(
                "get_state_and_qrid",
                arguments={
                    "latitude": "47.6062",
                    "longitude": "-122.3321"
                }
            )
            
            # Print the result
            print(f"\nTool Result:\n{result.content[0].text}")

# Run the client
await run_client()
```
